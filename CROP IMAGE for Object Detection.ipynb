```python
import numpy as np # linear algebra
import xml.etree.ElementTree as ET # for parsing XML
import matplotlib.pyplot as plt # to show images
from PIL import Image # to read images
import os
import glob
```

```python
root_images=r"C:\Users\Komol\Downloads\Autonomou Vehicle.v1i.voc\valid\\"
root_annots=r"C:\Users\Komol\Downloads\Autonomou Vehicle.v1i.voc\label_v\\"
```

```python
all_images=os.listdir(root_images)
print(f"Total images : {len(all_images)}")

breeds = glob.glob(r'C:\Users\Komol\Downloads\Autonomou Vehicle.v1i.voc\label_v\\')
annotation=[]
for b in breeds:
    annotation+=glob.glob(b+"/*")
print(f"Total annotation : {len(annotation)}")

breed_map={}
for annot in annotation:
    breed=annot.split("/")[-1]
    index=breed.split("-")[0]
    breed_map.setdefault(index,breed)
    
print(f"Total Breeds : {len(breed_map)}")
```

```python
def bounding_box(image):
    bpath=root_annots+str(image.split(".")[0]+".rf."+image.split(".")[2]+".xml")
    #root_annots+str(breed_map[image.split("_")[0]])+"/"+str(image.split(".")[0])
    print(bpath)
    tree = ET.parse(bpath)
    root = tree.getroot()
    objects = root.findall('object')
    for o in objects:
        bndbox = o.find('bndbox') # reading bound box
        if bndbox==bndbox:
            xmin = int(bndbox.find('xmin').text)
            ymin = int(bndbox.find('ymin').text)
            xmax = int(bndbox.find('xmax').text)
            ymax = int(bndbox.find('ymax').text)
        
        else:
            xmin = 0#int(bndbox.find('xmin').text)
            ymin = 0#int(bndbox.find('ymin').text)
            xmax = 0#int(bndbox.find('xmax').text)
            ymax = 0#int(bndbox.find('ymax').text)
        print(xmin)

    return (xmin,ymin,xmax,ymax)


def location(image):
    bpath=root_annots+str(image.split(".")[0]+".rf."+image.split(".")[2]+".xml")
    #root_annots+str(breed_map[image.split("_")[0]])+"/"+str(image.split(".")[0])
    #print(bpath)
    tree = ET.parse(bpath)
    root = tree.getroot()
    objects = root.findall('object')
    for o in objects:
        name=str(o.find('name').text)
        print(name)
    return (name)
```

```python
plt.figure(figsize=(10,10))
for i,image in enumerate(all_images):
    bbox=bounding_box(image)
    im=Image.open(os.path.join(root_images,image))
    im=im.crop(bbox)
    image_path = "E:/H Migrated One Drive all/D_DRIVE_OFFICE_PC/Scooter_YOLOv5/Cropped_Img_data/1000_train_data/val"
    image = im.save(f"{image_path}/"+location(image)+"/"+str(i)+".png")
    print(i)

    plt.subplot(120,120,i+1)
    plt.axis("off")
    plt.imshow(im)
    #os.mkdir(image_path)
    
    if(i==14400):
        break
```
