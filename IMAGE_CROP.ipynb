```python
import cv2, matplotlib.pyplot as plt, os
plt.ion()

#data_dir = "C:/STUDY_DRIVE_n10677984/Data\\train\\"
data_dir = "C:/Users/Komol/Komol_Deep_Transfer_Learning/data\\uncropped_abnormal\\"

for i in os.listdir(data_dir):
    temp_dir = data_dir + i + "\\"
    for j in os.listdir(temp_dir):
        img = cv2.imread(temp_dir + j)
        H, W, _ = img.shape
        cropped_img = img[40:H-68, 160:W-79]
        if not os.path.isdir(os.path.join("C:\\",'Users','Komol','Komol_Deep_Transfer_Learning','data','Pure_Training_Data','abnormal', i)) :
            os.mkdir(os.path.join("C:\\",'Users','Komol','Komol_Deep_Transfer_Learning','data','Pure_Training_Data','abnormal', i))
        print(cv2.imwrite(os.path.join("C:\\",'Users','Komol','Komol_Deep_Transfer_Learning','data','Pure_Training_Data','abnormal', i, j), cropped_img))
```
